## Ejercicio 1: Crear una clase abstracta

Una clase abstracta funciona como una plantilla.

En este caso, `ProductoBase` será una clase abstracta que define la estructura mínima que deben cumplir todos los productos.

Todo producto debe tener:

- nombre
- precio_base
- un método obligatorio llamado `calcular_precio_final()`

````python
class ProductoBase(ABC):
    def __init__(self, nombre, precio_base):
        self.nombre = nombre
        self.precio_base = precio_base

    @abstractmethod
    def calcular_precio_final(self):
        pass
````

## Verificación 1

Una clase abstracta no debe instanciarse directamente.

Ejecute la siguiente celda y observe el error controlado.

````python
try:
    producto = ProductoBase("Producto genérico", 100)
except TypeError as error:
    print("No se puede crear un objeto directamente desde una clase abstracta.")
    print("Error:", error)
````

## Ejercicio 2: Crear una clase hija ProductoFisico

La clase `ProductoFisico` debe heredar de `ProductoBase`.

Además de `nombre` y `precio_base`, tendrá:

- costo_envio

El precio final será:

precio final = precio base + costo de envío

````python
class ProductoFisico(ProductoBase):
    def __init__(self, nombre, precio_base, costo_envio):
        super().__init__(nombre, precio_base)
        self.costo_envio = costo_envio

    def calcular_precio_final(self):
        return self.precio_base + self.costo_envio
````

````python
producto1 = ProductoFisico("Teclado mecánico", 45, 5)

print(producto1.nombre)
print(producto1.precio_base)
print(producto1.costo_envio)
print(producto1.calcular_precio_final())
````

## Verificación 2

Compruebe que el precio final del producto físico se calcula correctamente.

````python
assert producto1.calcular_precio_final() == 50
print("Prueba superada: el precio final del producto físico es correcto.")
````

## Ejercicio 3: Crear una clase hija ProductoDigital

La clase `ProductoDigital` debe heredar de `ProductoBase`.

Además de `nombre` y `precio_base`, tendrá:

- descuento

El precio final será:

precio final = precio base - descuento

````python
class ProductoDigital(ProductoBase):
    def __init__(self, nombre, precio_base, descuento):
        super().__init__(nombre, precio_base)
        self.descuento = descuento

    def calcular_precio_final(self):
        return self.precio_base - self.descuento
````

````python
producto2 = ProductoDigital("Curso de Python", 30, 8)

print(producto2.nombre)
print(producto2.precio_base)
print(producto2.descuento)
print(producto2.calcular_precio_final())
````

## Ejercicio 5: Método especial `__str__`

El método `__str__` permite definir cómo se mostrará un objeto cuando se use `print()`.

Vamos a mejorar las clases para que los productos se impriman de forma más clara.

````python
class ProductoFisico(ProductoBase):
    def __init__(self, nombre, precio_base, costo_envio):
        super().__init__(nombre, precio_base)
        self.costo_envio = costo_envio

    def calcular_precio_final(self):
        return self.precio_base + self.costo_envio

    def __str__(self):
        return f"Producto físico: {self.nombre} | Precio final: ${self.calcular_precio_final()}"


class ProductoDigital(ProductoBase):
    def __init__(self, nombre, precio_base, descuento):
        super().__init__(nombre, precio_base)
        self.descuento = descuento

    def calcular_precio_final(self):
        return self.precio_base - self.descuento

    def __str__(self):
        return f"Producto digital: {self.nombre} | Precio final: ${self.calcular_precio_final()}"
````

````python
p1 = ProductoFisico("Laptop", 700, 25)
p2 = ProductoDigital("Licencia de software", 120, 20)

print(p1)
print(p2)
````

## Ejercicio 8: Patrón Singleton

El patrón Singleton permite que una clase tenga una única instancia.

En este ejercicio se creará una clase `ConfiguracionSistema`.

La idea es que todas las partes del programa usen la misma configuración.

````python
class ConfiguracionSistema:
    _instancia = None

    def __new__(cls):
        if cls._instancia is None:
            cls._instancia = super().__new__(cls)
            cls._instancia.modo = "desarrollo"
            cls._instancia.moneda = "USD"
        return cls._instancia

    def __str__(self):
        return f"Configuración: modo={self.modo}, moneda={self.moneda}"
````

````python
config1 = ConfiguracionSistema()
config2 = ConfiguracionSistema()

print(config1)
print(config2)

print("¿Son el mismo objeto?", config1 is config2)
````

````python
config1.modo = "producción"

print(config1)
print(config2)
````
