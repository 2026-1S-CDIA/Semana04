# ST Semana 4  
## Ejercicio 1: Sistema de personajes para un videojuego RPG

### Contexto

Un estudio de desarrollo de videojuegos desea crear un sistema básico para representar personajes de un juego RPG.

En el juego existen diferentes tipos de personajes, por ejemplo:

- Guerrero
- Mago
- Arquero

Todos los personajes tienen características comunes, pero cada tipo de personaje calcula su poder de ataque de manera diferente.

El objetivo del ejercicio es diseñar una solución usando Programación Orientada a Objetos en Python, aplicando:

- Abstracción
- Clases abstractas
- Herencia
- Polimorfismo
- Métodos especiales
- Sobrecarga de operadores
- Patrón Singleton

---

## Requerimientos generales

Debe crear una clase abstracta llamada `Personaje`.

Esta clase debe representar cualquier personaje del videojuego.

La clase `Personaje` debe tener los siguientes atributos:

- `nombre`
- `nivel`
- `vida`
- `energia`

Además, debe tener un método abstracto llamado:

- `calcular_poder_ataque()`

Este método no debe tener implementación en la clase abstracta, ya que cada tipo de personaje calculará su ataque de forma diferente.

---

## Clases hijas obligatorias

Debe crear tres clases hijas:

1. `Guerrero`
2. `Mago`
3. `Arquero`

Cada clase debe heredar de `Personaje`.

---

## Reglas de cálculo del poder de ataque

### Guerrero

El guerrero debe tener un atributo adicional:

- `fuerza`

El poder de ataque se calcula así:

poder = nivel * fuerza + energia * 0.2

---

### Mago

El mago debe tener un atributo adicional:

- `mana`

El poder de ataque se calcula así:

poder = nivel * mana + energia * 0.1

---

### Arquero

El arquero debe tener un atributo adicional:

- `precision`

El poder de ataque se calcula así:

poder = nivel * precision + energia * 0.15

---

## Métodos especiales obligatorios

Cada personaje debe implementar el método `__str__`.

Este método debe mostrar la información del personaje de forma clara.

Ejemplo esperado:

Personaje: Gandalf | Tipo: Mago | Nivel: 10 | Vida: 80 | Poder: 305.0

---

## Sobrecarga de operadores

Debe implementar los siguientes operadores:

### Operador suma `+`

Al sumar dos personajes, se debe obtener la suma de sus poderes de ataque.

Ejemplo:

personaje1 + personaje2

Debe devolver:

poder de ataque del personaje1 + poder de ataque del personaje2

---

### Operador mayor que `>`

Debe permitir comparar qué personaje tiene mayor poder de ataque.

Ejemplo:

personaje1 > personaje2

Debe devolver `True` si `personaje1` tiene mayor poder de ataque que `personaje2`.

---

### Operador igualdad `==`

Debe devolver `True` si dos personajes tienen el mismo poder de ataque.

---

## Patrón Singleton

Debe crear una clase llamada `ConfiguracionJuego`.

Esta clase debe aplicar el patrón Singleton.

La configuración debe almacenar:

- `nombre_juego`
- `dificultad`
- `version`

Valores iniciales:

- `nombre_juego`: "Python Quest"
- `dificultad`: "Normal"
- `version`: "1.0"

Debe comprobar que al crear dos objetos de configuración, ambos sean la misma instancia.

Ejemplo:

config1 = ConfiguracionJuego()
config2 = ConfiguracionJuego()

config1 is config2

Debe devolver:

True

---

## Pruebas obligatorias con `assert`

Debe crear al menos:

- 2 guerreros
- 2 magos
- 2 arqueros

Luego debe comprobar con `assert` que:

1. El poder de ataque de un guerrero se calcula correctamente.
2. El poder de ataque de un mago se calcula correctamente.
3. El poder de ataque de un arquero se calcula correctamente.
4. La suma de dos personajes devuelve la suma de sus poderes.
5. La comparación `>` funciona correctamente.
6. La comparación `==` funciona correctamente.
7. `ConfiguracionJuego` cumple con el patrón Singleton.

---

## Salida esperada

El programa debe imprimir un pequeño reporte como el siguiente:

=== CONFIGURACIÓN DEL JUEGO ===
Juego: Python Quest
Dificultad: Normal
Versión: 1.0

=== PERSONAJES CREADOS ===
Personaje: Thor | Tipo: Guerrero | Nivel: 8 | Vida: 100 | Poder: 168.0
Personaje: Merlin | Tipo: Mago | Nivel: 10 | Vida: 70 | Poder: 307.0
Personaje: Legolas | Tipo: Arquero | Nivel: 9 | Vida: 85 | Poder: 232.5

=== COMPARACIONES ===
Merlin tiene mayor poder que Thor: True
Thor y Legolas tienen el mismo poder: False
Poder combinado de Thor y Merlin: 475.0

---

## Preguntas de análisis

Responda al final del ejercicio:

1. ¿Por qué `Personaje` debe ser una clase abstracta?
2. ¿Dónde se aplica el polimorfismo?
3. ¿Qué ventaja tiene sobrecargar el operador `+` en este caso?
4. ¿Qué ventaja tiene sobrecargar el operador `>`?
5. ¿Por qué `ConfiguracionJuego` puede implementarse como Singleton?
6. ¿Qué problema podría aparecer si se abusa del patrón Singleton?

# ST Semana 4  
## Ejercicio 2: Sistema de vehículos para una carrera futurista

### Contexto

Una empresa está desarrollando un simulador de carreras futuristas. En este simulador existen distintos tipos de vehículos:

- Auto eléctrico
- Moto futurista
- Dron de carrera

Todos los vehículos participan en competencias, pero cada uno calcula su rendimiento de forma diferente.

El sistema debe permitir registrar vehículos, calcular su rendimiento, comparar cuál es mejor y administrar una configuración global de la carrera.

El ejercicio debe resolverse aplicando Programación Orientada a Objetos en Python.

---

## Conceptos obligatorios

En este ejercicio debe aplicar:

- Abstracción
- Clase abstracta
- Herencia
- Polimorfismo
- Método `__str__`
- Sobrecarga de operadores
- Patrón Singleton
- Pruebas con `assert`

---

## Clase abstracta

Debe crear una clase abstracta llamada `VehiculoCarrera`.

Esta clase debe tener los siguientes atributos:

- `nombre`
- `velocidad_maxima`
- `peso`
- `energia`

La clase debe tener un método abstracto llamado:

- `calcular_rendimiento()`

Este método representa qué tan competitivo es un vehículo.

No debe implementarse directamente en `VehiculoCarrera`.

---

## Clases hijas

Debe crear tres clases hijas:

1. `AutoElectrico`
2. `MotoFuturista`
3. `DronCarrera`

Cada una debe heredar de `VehiculoCarrera`.

---

## Reglas de rendimiento

### AutoElectrico

Debe tener un atributo adicional:

- `capacidad_bateria`

El rendimiento se calcula así:

rendimiento = velocidad_maxima * 0.6 + capacidad_bateria * 0.3 - peso * 0.05 + energia * 0.1

---

### MotoFuturista

Debe tener un atributo adicional:

- `maniobrabilidad`

El rendimiento se calcula así:

rendimiento = velocidad_maxima * 0.5 + maniobrabilidad * 0.4 - peso * 0.03 + energia * 0.1

---

### DronCarrera

Debe tener un atributo adicional:

- `estabilidad_vuelo`

El rendimiento se calcula así:

rendimiento = velocidad_maxima * 0.55 + estabilidad_vuelo * 0.35 - peso * 0.02 + energia * 0.1

---

## Método `__str__`

Cada vehículo debe poder imprimirse de forma clara.

Ejemplo esperado:

Vehículo: Falcon X | Tipo: DronCarrera | Velocidad: 220 km/h | Rendimiento: 174.6

---

## Sobrecarga de operadores

Debe implementar los siguientes operadores:

### Operador suma `+`

Al sumar dos vehículos, debe devolver la suma de sus rendimientos.

Ejemplo:

vehiculo1 + vehiculo2

---

### Operador resta `-`

Al restar dos vehículos, debe devolver la diferencia entre sus rendimientos.

Ejemplo:

vehiculo1 - vehiculo2

---

### Operador menor que `<`

Debe permitir comparar si un vehículo tiene menor rendimiento que otro.

Ejemplo:

vehiculo1 < vehiculo2

---

## Clase Singleton

Debe crear una clase llamada `ConfiguracionCarrera`.

Esta clase debe aplicar el patrón Singleton.

Debe almacenar:

- `nombre_torneo`
- `ciudad`
- `numero_vueltas`
- `modo`

Valores iniciales:

- `nombre_torneo`: "Neo Grand Prix"
- `ciudad`: "Quito Futurista"
- `numero_vueltas`: 5
- `modo`: "Clasificatoria"

Debe implementar `__str__`.

---

## Comprobación del Singleton

Debe crear dos objetos:

config1 = ConfiguracionCarrera()
config2 = ConfiguracionCarrera()

Luego debe comprobar:

config1 is config2

El resultado debe ser:

True

Después, cambie el número de vueltas desde `config1`.

Ejemplo:

config1.numero_vueltas = 8

Al imprimir `config2`, también debe aparecer el número de vueltas actualizado.

---

## Escenario mínimo de prueba

Debe crear al menos seis vehículos:

- 2 autos eléctricos
- 2 motos futuristas
- 2 drones de carrera

Luego debe almacenarlos en una lista llamada `vehiculos`.

Debe recorrer la lista e imprimir cada vehículo.

También debe identificar cuál tiene el mayor rendimiento.

---

## Pruebas obligatorias con `assert`

Debe comprobar con `assert` que:

1. El rendimiento de un auto eléctrico se calcula correctamente.
2. El rendimiento de una moto futurista se calcula correctamente.
3. El rendimiento de un dron de carrera se calcula correctamente.
4. La suma de dos vehículos funciona con `+`.
5. La resta de dos vehículos funciona con `-`.
6. La comparación `<` funciona correctamente.
7. La clase `ConfiguracionCarrera` cumple Singleton.
8. Al modificar `config1`, el cambio se refleja en `config2`.

---

## Reporte esperado

El programa debe imprimir un reporte similar al siguiente:

=== CONFIGURACIÓN DE CARRERA ===
Torneo: Neo Grand Prix
Ciudad: Quito Futurista
Vueltas: 5
Modo: Clasificatoria

=== VEHÍCULOS REGISTRADOS ===
Vehículo: Volt Runner | Tipo: AutoElectrico | Velocidad: 250 km/h | Rendimiento: 183.5
Vehículo: Shadow Bike | Tipo: MotoFuturista | Velocidad: 230 km/h | Rendimiento: 176.2
Vehículo: Falcon X | Tipo: DronCarrera | Velocidad: 220 km/h | Rendimiento: 174.6

=== RESULTADOS ===
Vehículo con mayor rendimiento: Volt Runner
Rendimiento combinado de Volt Runner y Falcon X: 358.1
Diferencia entre Volt Runner y Shadow Bike: 7.3

---

## Restricciones

1. No debe crear objetos directamente desde `VehiculoCarrera`.
2. Cada clase hija debe implementar obligatoriamente `calcular_rendimiento()`.
3. No debe repetir innecesariamente código.
4. Debe usar `super()` en las clases hijas.
5. Los operadores deben usar el método `calcular_rendimiento()`, no atributos directos.
6. El código debe estar organizado y comentado brevemente.

---

3. ¿Dónde se evidencia el polimorfismo?
4. ¿Qué ventaja tiene usar `__str__`?
5. ¿Qué ventaja tiene sobrecargar `+`, `-` y `<`?
